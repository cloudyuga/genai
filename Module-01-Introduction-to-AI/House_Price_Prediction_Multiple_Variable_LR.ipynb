```python
import pandas as pd
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/cloudyuga/genai/main/Module-01-Introduction-to-AI/train.csv")
```

```python
dataset
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X = dataset[["size_sq_feet", "kms_from_metro"]]
```

```python
y = dataset["price_in_lacs_inr"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print ('Residual sum of squares: %.2f' % np.mean((model.predict(X_train)- y_train) ** 2))
```

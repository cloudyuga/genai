##**1. Import Statements**

```python
import os
import tempfile
import gradio as gr
from openai import OpenAI
from pathlib import Path
```

##**2. Create OpenAI client to use whisper model**

```python
from langchain_core.tools import tool
from google.colab import userdata
# Retrieve the API key from Colab's secrets
api_key = userdata.get('openai_api_key')
```

```python
client = OpenAI(api_key=api_key)
```

##**3. Function to Translate Audio**

```python
def translate_audio(tmp_filepath: str) -> str:
    """
    Translates an audio file to English via OpenAI Whisper.

    Args:
        tmp_filepath: Path to the temporary audio file recorded by Gradio.
        response_format: "text" for plain translation or "verbose_json" for detailed segments.
    """
    if not tmp_filepath:
        print("No audio recorded.")
        return "No audio recorded."

    with open(tmp_filepath, "rb") as audio_file:
        result = client.audio.translations.create(
            model="whisper-1",
            file=audio_file,
            response_format= "text"
        )
    return result
```

##**4. Build Gradio UI**

```python
# Build Gradio interface for translation only
iface = gr.Interface(
    fn=translate_audio,
    inputs=[
        gr.Audio(label="Your Recording", type="filepath"),
    ],
    outputs=gr.Textbox(label="Result", lines=15),
    title="ðŸŽ¤ Whisper Translation",
    description="Record your voice and get an English translation via OpenAI Whisper.",
    allow_flagging="never"
)

if __name__ == "__main__":
    iface.launch()
```
